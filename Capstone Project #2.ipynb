## Step 1: Import Libraries

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

## Step 2: Import Advertising Data

```python
advertising_data=pd.read_csv('Advertising_Data.csv')
```

```python
print(advertising_data.head())
print(advertising_data.describe())
print(advertising_data.info())
```

## Step 3: Cleanse the Data

```python
advertising_data = advertising_data.drop_duplicates()
advertising_data = advertising_data.dropna()
advertising_data = advertising_data.drop(columns = "Billboards")
print(advertising_data.info())
```

Above, you can see a few lines of code meant to check the cleanliness of the dataset. The first line dropped all duplicate entries to make sure duplicate information wouldn't skew or disrupt the dataset. The second line, removed any null values from the dataset, which would be a data antry error. 

We also decided to remove the "Billboard" column because our client does there business 100% online, and has clearly stated they are not interested in billboard advertising. Also, I printed the info of the dataset to see if these changes significantly cleaned the datase. There isn't a significant difference to our dataset whn you compare the "info" results before and after our code written to cleanse the data. This means that the dataset has a high level of cleanliness!

## Step 4: Analyze the Data


Next, we will analyze the data to see if there are actionalble insights in the datasets that could be useful to the team at Innovtech. I want to start with analyzing the amount of correlation between the product sold and each of the other quantitative variables in the dataset. 

```python
g = sns.relplot(x="TV",
           y="Product_Sold",
           data=advertising_data,
           kind="scatter")
g.fig.suptitle('TV Advertisements', fontsize=14)
plt.show()
```

```python
g = sns.relplot(x="Google_Ads",
           y="Product_Sold",
           data=advertising_data,
           kind="scatter")
g.fig.suptitle('Google Advertisements', fontsize=14)
plt.show()
```

```python
g = sns.relplot(x="Social_Media",
           y="Product_Sold",
           data=advertising_data,
           kind="scatter")
g.fig.suptitle('Social Media', fontsize=14)
plt.show()
```

```python
g = sns.relplot(x="Influencer_Marketing",
           y="Product_Sold",
           data=advertising_data,
           kind="scatter")
g.fig.suptitle('Influencer_Marketing', fontsize=14)
plt.show()
```

```python
g = sns.relplot(x="Affiliate_Marketing",
           y="Product_Sold",
           data=advertising_data,
           kind="scatter")
g.fig.suptitle('Affiliate Marketing', fontsize=14)
plt.show()
```

After, creating these scatterplot graphs it is clear that there are varying levels or correlation. Therefore, to further analyze the data I would like to add a reference line to show viewers the slope of the correlation. 

After I add the reference lines, I will calculate the correlation of all the visualizations shown above.
